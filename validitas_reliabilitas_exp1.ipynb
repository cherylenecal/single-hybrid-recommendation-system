```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
df = pd.read_excel("/content/Kuesioner Kepribadian (Responses).xlsx")
```

```python
df.columns = df.columns.str.extract(r'\((\w+-\w+\d+)\)').iloc[:, 0]
display(df.head())
```

```python
import numpy as np

def cronbach_alpha(df):
    """Calculates Cronbach's alpha for a DataFrame."""
    # Number of items
    n_items = df.shape[1]

    # Calculate the variance of each item
    item_variances = df.var(axis=0, ddof=1)

    # Calculate the variance of the sum of the items
    total_variance = df.sum(axis=1).var(ddof=1)

    # Calculate Cronbach's alpha
    alpha = (n_items / (n_items - 1)) * (1 - (item_variances.sum() / total_variance))

    return alpha

# Define the groups of columns to analyze
groups = {
    'INN': ['INN-CE1', 'INN-CE2', 'INN-O1', 'INN-O2', 'INN-W1', 'INN-W2'],
    'SEM': ['SE-M1', 'SE-M2'],
    'SEP': ['SE-P1', 'SE-P2'],
    'SEIP': ['SE-IP1', 'SE-IP2'],
    'SEIF': ['SE-IF1', 'SE-IF2'],
    'SES': ['SE-S1', 'SE-S2'],
    'LOCI': ['LOC-I1', 'LOC-I2'],
    'LOCE': ['LOC-E1', 'LOC-E2'],
    'NACHFF': ['NACH-FF1', 'NACH-FF2'],
    'NACHHS': ['NACH-HS1', 'NACH-HS2']
}

# Calculate and display Cronbach's alpha for each group
reliability_results = {}
for group_name, columns in groups.items():
    if all(col in df.columns for col in columns):
        reliability_results[group_name] = cronbach_alpha(df[columns])
    else:
        reliability_results[group_name] = f"One or more columns not found: {columns}"


for group, alpha in reliability_results.items():
    print(f"Cronbach's Alpha for {group}: {alpha}")
```

```python
# 4. Threshold Validitas (N=30 -> 0.361)
r_table = 0.361

# 5. Loop Perhitungan Validitas
print("\n--- HASIL UJI VALIDITAS (Item-Total Pearson Correlation) ---")
for g_name, cols in groups.items():
    print(f"\nVariabel: {g_name}")

    subset = df[cols]
    total_score = subset.sum(axis=1)

    v_results = []
    for col in cols:
        r, p = pearsonr(subset[col], total_score)
        v_results.append({
            "Item": col,
            "Pearson r": round(float(r), 3),
            "p-value": round(float(p), 4),
            "Status": "Valid" if float(r) > r_table else "Tidak Valid"
        })

    df_validitas = pd.DataFrame(v_results)
    print(df_validitas)
```
